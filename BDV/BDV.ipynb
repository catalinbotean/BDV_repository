```python
import requests

# GDP growth (annual %)
def getCountryData(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/NY.GDP.MKTP.KD.ZG?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'GDP Growth': d['value'] or 0} for d in data]

# GDP per capita growth (annual %)
def getCountryGDPCapital(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.KD.ZG?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'GDP per Capita': d['value'] or 0} for d in data]

# Population ages 65 and above (% of total population)
def getMortalityData(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.POP.65UP.TO.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Population over 65y': d['value'] or 0} for d in data]

# Access to electricity (% of population)
def getAccessToElectricity(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.ELC.ACCS.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Access To Electricity': d['value'] or 0} for d in data]

# Forest area (% of land area)
def getForestAreaPercentage(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/AG.LND.FRST.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Forest Area Percentage': d['value'] or 0} for d in data]

# Forest area (sq. km)
def getForestArea(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/AG.LND.FRST.K2?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Forest Area': d['value'] or 0} for d in data]

# Rural population
def getRuralPopulation(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.RUR.TOTL?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Rural Population': d['value'] or 0} for d in data]

# Rural population (% of total population)
def getRuralPopulationPercentage(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.RUR.TOTL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Rural Population Percentage': d['value'] or 0} for d in data]

# Urban population
def getUrbanPopulation(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.URB.TOTL?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Urban Population': d['value'] or 0} for d in data]

# Urban population (% of total population)
def getUrbanPopulationPercentage(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.URB.TOTL.IN.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Urban Population Percentage': d['value'] or 0} for d in data]

# CO2 emissions (metric tons per capita)
def getCO2Emissions(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.ATM.CO2E.PC?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'CO2': d['value'] or 0} for d in data]

# CO2 emissions (kt)
def getCO2kt(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.ATM.CO2E.KT?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'CO2 kt': d['value'] or 0} for d in data]

# Energy use (kg of oil equivalent per capita)
def getEnergyUse(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.USE.PCAP.KG.OE?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Energy Use': d['value'] or 0} for d in data]

# Electric power consumption (kWh per capita)
def getElectricPowerConsumption(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.USE.ELEC.KH.PC?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Electric Power Consumption': d['value'] or 0} for d in data]


# Methane emissions (kt of CO2 equivalent)
def getMethaneEmissions(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.ATM.METH.KT.CE?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Methane Emissions': d['value'] or 0} for d in data]

# Nitrous oxide emissions (thousand metric tons of CO2 equivalent)
def getNOEmissions(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.ATM.NOXE.KT.CE?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'N2O Emissions': d['value'] or 0} for d in data]

# Population in urban agglomerations of more than 1 million (% of total population)
def getPopulationInUrbanAgglomeration(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.URB.MCTY.TL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Urban Agglomeration': d['value'] or 0} for d in data]

# Population growth (annual %)
def getPopulationGrowth(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.POP.GROW?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Population Growth': d['value'] or 0} for d in data]

# Population, total
def getPopulation(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Population': d['value'] or 0} for d in data]

# Renewable energy consumption (% of total final energy consumption)
def getRenewableEnergyConsumption(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.FEC.RNEW.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'RenewableEnergyConsumption': d['value'] or 0} for d in data]

# Terrestrial and marine protected areas (% of total territorial area)
def getProtectedAreas(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/ER.PTD.TOTL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Protected Areas': d['value'] or 0} for d in data]

# Total greenhouse gas emissions (kt of CO2 equivalent)
def getGreenhouseGas(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EN.ATM.GHGT.KT.CE?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Greenhouse Gas': d['value'] or 0} for d in data]

# Industry (including construction), value added (% of GDP)
def getIndustryValue(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/NV.IND.TOTL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Industry Value': d['value'] or 0} for d in data]

# Inflation, consumer prices (annual %)
def getInflation(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/FP.CPI.TOTL.ZG?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Inflation': d['value'] or 0} for d in data]

# Imports of goods and services (% of GDP)
def getImports(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/NE.IMP.GNFS.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Imports': d['value'] or 0} for d in data]

# Agriculture, forestry, and fishing, value added (% of GDP)
def getAgricultureForestryFishing(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/NV.AGR.TOTL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Agriculture Forestry Fishing': d['value'] or 0} for d in data]

# Fossil fuel energy consumption (% of total)
def getFossilFuelEnergyConsumption(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.USE.COMM.FO.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Fossil Fuel': d['value'] or 0} for d in data]

# Ores and metals exports (% of merchandise exports)
def getMetalExports(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/TX.VAL.MMTL.ZS.UN?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Metal Exports': d['value'] or 0} for d in data]

# Alternative and nuclear energy (% of total energy use)
def getAlternativeEnergy(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.USE.COMM.CL.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Alternative Energy': d['value'] or 0} for d in data]

# Energy imports, net (% of energy use)
def getEnergyImports(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.IMP.CONS.ZS?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Energy Imports': d['value'] or 0} for d in data]

# Energy use (kg of oil equivalent per capita)
def getEnergyUse(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/EG.USE.PCAP.KG.OE?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Energy Use': d['value'] or 0} for d in data]

# Fuel exports (% of merchandise exports)
def getFuelExports(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/TX.VAL.FUEL.ZS.UN?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Fuel Exports': d['value'] or 0} for d in data]

# Air transport, registered carrier departures worldwide
def getAirTransport(year):
    response = requests.get(f"https://api.worldbank.org/v2/country/all/indicator/IS.AIR.DPRT?date={year}&format=json&per_page=5000")
    data = response.json()[1]
    return [{'Country': d['country']['value'], 'Air Transport': d['value'] or 0} for d in data]
```

```python
import pandas as pd

def createDataset(year):
  # Fetching country GDP data for the year 2020
  countryData = getCountryData(year)

  # Fetching country GDP Capital data for the year 2020
  gdpCapitalData = getCountryGDPCapital(year)

  # Fetching country Mortality (people over 65 years) data for the year 2020
  mortalityData = getMortalityData(year)

  # Fetching country Access To Electricity data for the year 2020
  accessToElectricityData = getAccessToElectricity(year)

  # Fetching country Forest Area Percentage data for the year 2020
  forestAreaPercentage = getForestAreaPercentage(year) 

  # Fetching country Forest Area data for the year 2020
  forestArea = getForestArea(year) 

  # Fetching country Rural population data data for the year 2020
  ruralPopulation = getRuralPopulation(year) 

  # Fetching country Rural population percentage data data for the year 2020
  ruralPopulationPercentage = getRuralPopulationPercentage(year) 

  # Fetching country Urban population data data for the year 2020
  urbanPopulation = getUrbanPopulation(year) 

  # Fetching country Urban population percentage data data for the year 2020
  urbanPopulationPercentage = getUrbanPopulationPercentage(year) 

  # Fetching country CO2 emissions data data for the year 2020
  co2Emissions = getCO2Emissions(year)

  # Fetching country CO2 emissions kt data data for the year 2020
  co2EmissionsKT = getCO2kt(year)

  # Fetching country energy use data data for the year 2020
  energyUse = getEnergyUse(year)

  # Fetching country electric power consumption data data for the year 2020
  electriPowerConsumption = getElectricPowerConsumption(year)

  # Fetching country Methane emissions data data for the year 2020
  methaneEmissions = getMethaneEmissions(year)

  # Fetching country nitrous oxide emissions data data for the year 2020
  nitrousOxideEmissions = getNOEmissions(year)

  # Fetching country Population In Urban Agglomeration data data for the year 2020
  populationInUrbanAgglomeration = getPopulationInUrbanAgglomeration(year)

  # Fetching country Population Growth data data for the year 2020
  populationGrowth = getPopulationGrowth(year)

  # Fetching country Population data for the year 2020
  population = getPopulation(year)

  # Fetching country Renewable energy consumption data for the year 2020
  renewableEnergyConsumption = getRenewableEnergyConsumption(year)

  # Fetching country Protected areas data for the year 2020
  protectedAreas = getProtectedAreas(year)

  # Fetching country Greenhouse gas data for the year 2020
  greenhouseGas = getGreenhouseGas(year)

  # Fetching country industry value data for the year 2020
  industryValue = getIndustryValue(year)

  # Fetching country inflation data for the year 2020
  inflation = getInflation(year)

  # Fetching country imports data for the year 2020
  imports = getImports(year)

  # Fetching country Agriculture Forestry Fishing data for the year 2020
  agricultureForestryFishing = getAgricultureForestryFishing(year)

  # Fetching country fossil fuel data for the year 2020
  fossilFuel = getFossilFuelEnergyConsumption(year)

  # Fetching country metal exports data for the year 2020
  metalExports = getMetalExports(year)

  # Fetching country alternative energy data for the year 2020
  alternativeEnergy = getAlternativeEnergy(year)

  # Fetching country energy imports data for the year 2020
  energyImport = getAlternativeEnergy(year)

  # Fetching country energy use data for the year 2020
  energyUse = getEnergyUse(year)

  # Fetching country fuel exports data for the year 2020
  fuelExports = getFuelExports(year)

  # Fetching country air transport data for the year 2020
  airTransport = getAirTransport(year)

  df = pd.DataFrame(countryData)
  df = pd.merge(df, pd.DataFrame(gdpCapitalData), on='Country')
  df = pd.merge(df, pd.DataFrame(mortalityData), on='Country')
  df = pd.merge(df, pd.DataFrame(accessToElectricityData), on='Country')
  df = pd.merge(df, pd.DataFrame(forestAreaPercentage), on='Country')
  df = pd.merge(df, pd.DataFrame(ruralPopulation), on='Country')
  df = pd.merge(df, pd.DataFrame(ruralPopulationPercentage), on='Country')
  df = pd.merge(df, pd.DataFrame(urbanPopulation), on='Country')
  df = pd.merge(df, pd.DataFrame(urbanPopulationPercentage), on='Country')
  df = pd.merge(df, pd.DataFrame(co2Emissions), on='Country')
  df = pd.merge(df, pd.DataFrame(co2EmissionsKT), on='Country')
  df = pd.merge(df, pd.DataFrame(energyUse), on='Country')
  df = pd.merge(df, pd.DataFrame(electriPowerConsumption), on='Country')
  df = pd.merge(df, pd.DataFrame(methaneEmissions), on='Country')
  df = pd.merge(df, pd.DataFrame(nitrousOxideEmissions), on='Country')
  df = pd.merge(df, pd.DataFrame(populationInUrbanAgglomeration), on='Country')
  df = pd.merge(df, pd.DataFrame(populationGrowth), on='Country')
  df = pd.merge(df, pd.DataFrame(renewableEnergyConsumption), on='Country')
  df = pd.merge(df, pd.DataFrame(protectedAreas), on='Country')
  df = pd.merge(df, pd.DataFrame(greenhouseGas), on='Country')
  df = pd.merge(df, pd.DataFrame(industryValue), on='Country')
  df = pd.merge(df, pd.DataFrame(inflation), on='Country')
  df = pd.merge(df, pd.DataFrame(imports), on='Country')
  df = pd.merge(df, pd.DataFrame(agricultureForestryFishing), on='Country')
  df = pd.merge(df, pd.DataFrame(fossilFuel), on='Country')
  df = pd.merge(df, pd.DataFrame(metalExports), on='Country')
  df = pd.merge(df, pd.DataFrame(alternativeEnergy), on='Country')
  df = pd.merge(df, pd.DataFrame(energyImport), on='Country')
  df = pd.merge(df, pd.DataFrame(energyUse), on='Country')
  df = pd.merge(df, pd.DataFrame(fuelExports), on='Country')
  df = pd.merge(df, pd.DataFrame(airTransport), on='Country')
  df['Year'] = year
  cols = list(df.columns)
  cols = [cols[0]] + [cols[len(cols)-1]] + cols[1:len(cols)-1]
  df = df.reindex(columns=cols)
  return df
```

```python
dfs = []
for year in range(1960, 2023):
    df = createDataset(year)
    dfs += [df]
df = pd.concat(dfs)
```
